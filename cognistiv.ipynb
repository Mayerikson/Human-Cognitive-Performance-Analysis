#  Justificativa para a Altera√ß√£o do Tipo de Dados de `Caffeine_Intake`

## Contexto  
O dataset `"human_cognitive_performance.csv"` cont√©m a vari√°vel `Caffeine_Intake`, que representa o consumo di√°rio de cafe√≠na em miligramas (mg) por usu√°rio. Os metadados indicam que essa vari√°vel deveria ser tratada como um valor num√©rico cont√≠nuo, permitindo maior granularidade e precis√£o.

---

## Tipo de Dados Atual  
- **Tipo Observado:** `int64` (inteiro de 64 bits).  
- **Baseado nos Resultados:** A contagem de tipos de dados (`dtypes`) mostrou que `Caffeine_Intake` est√° armazenada como `int64`, o que restringe os valores a n√∫meros inteiros (ex.: 180, 302, 337).  

---

## Tipo de Dados Esperado  
- **Tipo Esperado:** `float64` (ponto flutuante de 64 bits).  
- **Justificativa:**  
  - O consumo de cafe√≠na pode incluir valores decimais (ex.: 180.5 mg), especialmente em contextos reais onde bebidas ou suplementos podem ser medidos com precis√£o fracion√°ria.  
  - O uso de `float64` √© mais apropriado para vari√°veis cont√≠nuas, alinhando-se com a natureza potencialmente n√£o inteira de `Caffeine_Intake`, conforme sugerido pela descri√ß√£o no dicion√°rio de dados ("mg por dia").  
  - Isso evita a perda de informa√ß√£o e permite an√°lises estat√≠sticas mais precisas, como m√©dias ponderadas ou regress√µes que dependem de valores exatos.  

---

## Impacto da Altera√ß√£o  
- **Benef√≠cio:** Converter `Caffeine_Intake` para `float64` garante que o dataset suporte valores decimais, refletindo melhor a realidade dos dados e melhorando a robustez das an√°lises preditivas (ex.: Random Forest Regressor e regress√£o m√∫ltipla linear).  
- **Risco:** N√£o h√° risco significativo, pois a convers√£o de inteiros para ponto flutuante preserva os valores originais (ex.: 180 torna-se 180.0), e o dataset n√£o cont√©m valores ausentes (`isnull().sum() = 0`).  

# 4 Explora√ß√£o Gr√°fica inicial

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar estilo do seaborn (para consist√™ncia com gr√°ficos anteriores)
sns.set_style("whitegrid")
plt.rcParams['axes.titlesize'] = 12
plt.rcParams['axes.labelsize'] = 10

# Passo 3: Criar vari√°veis derivadas relevantes para as perguntas de neg√≥cio
# 3.1. Diferen√ßa de desempenho cognitivo (Memory_Test_Score - AI_Predicted_Score)
data['Cognitive_Difference'] = data['Memory_Test_Score'] - data['AI_Predicted_Score']

# 3.2. Taxa de erro preditivo (para an√°lise de precis√£o do modelo)
data['Prediction_Error_Rate'] = (abs(data['Cognitive_Difference']) / data['Memory_Test_Score']) * 100

# 3.3. √çndice de estresse ajustado por sono (para pergunta sobre sono e estresse)
data['Stress_Sleep_Index'] = data['Stress_Level'] / (data['Sleep_Duration'] + 0.01)  # +0.01 para evitar divis√£o por zero

# Passo 4: Verificar outliers nas vari√°veis quantitativas (m√©todo IQR)
print("\nVerifica√ß√£o de Outliers (M√©todo IQR)")
quantitative_vars = [
    'Age', 'Sleep_Duration', 'Stress_Level', 'Daily_Screen_Time', 'Caffeine_Intake',
    'Reaction_Time', 'Memory_Test_Score', 'Cognitive_Score', 'AI_Predicted_Score',
    'Cognitive_Difference', 'Prediction_Error_Rate', 'Stress_Sleep_Index'
]

for col in quantitative_vars:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)][col]
    print(f"\nColuna: {col}")
    print(f"Limite Inferior: {lower_bound:.2f}, Limite Superior: {upper_bound:.2f}")
    print(f"N√∫mero de Outliers: {len(outliers)}")
    if len(outliers) > 0:
        print(f"Valores dos Outliers (at√© 10): {outliers.tolist()[:10]}")

# Passo 5: Padronizar e criar vari√°veis categ√≥ricas
# 5.1. Criar Cognitive_Performance_Category (baixo, m√©dio, alto) baseado em Memory_Test_Score
print("\nCria√ß√£o de Cognitive_Performance_Category")
data['Cognitive_Performance_Category'] = pd.cut(data['Memory_Test_Score'],
                                             bins=[0, data['Memory_Test_Score'].quantile(0.33),
                                                   data['Memory_Test_Score'].quantile(0.66),
                                                   data['Memory_Test_Score'].max()],
                                             labels=['Baixo', 'M√©dio', 'Alto'],
                                             include_lowest=True)
print("Valores √∫nicos de Cognitive_Performance_Category:", data['Cognitive_Performance_Category'].unique())

# 5.2. Criar High_Stress_Level (bin√°rio: 1 para Stress_Level > 7, 0 para <= 7)
print("\nCria√ß√£o de High_Stress_Level")
data['High_Stress_Level'] = (data['Stress_Level'] > 7).astype(int)
print("Valores √∫nicos de High_Stress_Level:", data['High_Stress_Level'].unique())

# 5.3. Criar Sleep_Quality_Level
print("\nCria√ß√£o de Sleep_Quality_Level")
data['Sleep_Quality_Level'] = pd.cut(data['Sleep_Duration'],
                                   bins=[0, 6, 8, data['Sleep_Duration'].max()],
                                   labels=['Baixo (<6h)', 'M√©dio (6-8h)', 'Alto (>8h)'])
print("Valores √∫nicos de Sleep_Quality_Level:", data['Sleep_Quality_Level'].unique())

# Passo 6: Verificar valores √∫nicos das vari√°veis qualitativas
print("\nValores √∫nicos das vari√°veis qualitativas:")
qualitative_vars = ['Gender', 'Diet_Type', 'Exercise_Frequency']
for var in qualitative_vars:
    print(f"\n{var}:")
    print(data[var].unique())

# Passo 7: Exibir resumo do DataFrame atualizado
print("\nResumo do DataFrame ap√≥s Corre√ß√µes")
print(data.info())
print("\nPrimeiras 5 Linhas do DataFrame Atualizado")
print(data.head())
````

### Defini√ß√£o do M√©todo IQR
Os limites foram calculados como:
- **Limite inferior**: $ Q1 - 1.5 \times IQR $
- **Limite superior**: $ Q3 + 1.5 \times IQR $

Onde:
- $ Q1 $ = primeiro quartil
- $ Q3 $ = terceiro quartil  
- $ IQR = Q3 - Q1 $

# 6 Respondendo as Perguntas de Neg√≥cio

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr, ttest_ind
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import shap
import warnings
warnings.filterwarnings('ignore')

class CompleteCognitiveAnalysis:
    """
    Classe completa para an√°lise de desempenho cognitivo com otimiza√ß√µes
    """
    
    def __init__(self, data, target_col='Cognitive_Score'):
        self.raw_data = data.copy()
        self.target_col = target_col
        self.label_encoders = {}
        self.data = self._preprocess_data(data.copy())
        self.scaler = StandardScaler()
        self.results = {}
        
    def _preprocess_data(self, data):
        """Pr√©-processamento otimizado"""
        # Remover colunas n√£o relevantes
        cols_to_drop = ['User_ID', 'AI_Predicted_Score', 'Cognitive_Difference', 
                       'Prediction_Error_Rate', 'Stress_Sleep_Index']
        data = data.drop(columns=[col for col in cols_to_drop if col in data.columns])
        
        # Converter colunas categ√≥ricas
        categorical_cols = ['Gender', 'Diet_Type', 'Exercise_Frequency', 
                          'Cognitive_Performance_Category', 'Sleep_Quality_Level']
        
        for col in categorical_cols:
            if col in data.columns:
                if data[col].dtype.name == 'category':
                    data[col] = data[col].astype('object')
                le = LabelEncoder()
                data[col] = le.fit_transform(data[col].astype(str))
                self.label_encoders[col] = le
        
        # Tratar valores ausentes
        numeric_cols = data.select_dtypes(include=[np.number]).columns
        data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].median())
        
        return data

    def answer_question1(self):
        """Pergunta 1: Fatores associados ao bom desempenho"""
        print("\n1. QUAIS FATORES EST√ÉO MAIS ASSOCIADOS AO BOM DESEMPENHO COGNITIVO?")
        
        numeric_cols = ['Age', 'Sleep_Duration', 'Stress_Level', 'Daily_Screen_Time',
                       'Caffeine_Intake', 'Reaction_Time', 'Memory_Test_Score',
                       'High_Stress_Level']
        
        available_cols = [col for col in numeric_cols if col in self.data.columns]
        
        # Calcular correla√ß√µes
        correlations = {}
        for col in available_cols:
            corr, p_value = spearmanr(self.data[self.target_col], self.data[col])
            if not np.isnan(corr):
                correlations[col] = {
                    'correlation': corr,
                    'p_value': p_value,
                    'strength': self._interpret_correlation(abs(corr))
                }
        
        # Ordenar por correla√ß√£o absoluta
        sorted_corrs = sorted(correlations.items(), 
                            key=lambda x: abs(x[1]['correlation']), 
                            reverse=True)
        
        # Resultados
        print("\nTop 5 correla√ß√µes com desempenho cognitivo:")
        for col, stats in sorted_corrs[:5]:
            sig = "***" if stats['p_value'] < 0.001 else "**" if stats['p_value'] < 0.01 else "*" if stats['p_value'] < 0.05 else ""
            print(f"{col}: r = {stats['correlation']:.2f}{sig} ({stats['strength']})")
        
        # Visualiza√ß√£o
        top_features = [col for col, _ in sorted_corrs[:3]]
        if top_features:
            self._plot_correlations(top_features)
        
        self.results['question1'] = correlations
        return correlations
    
    def _interpret_correlation(self, r):
        """Interpreta a for√ßa da correla√ß√£o"""
        if r >= 0.7: return "Muito Forte"
        elif r >= 0.5: return "Forte"
        elif r >= 0.3: return "Moderada"
        elif r >= 0.1: return "Fraca"
        return "Desprez√≠vel"
    
    def _plot_correlations(self, features):
        """Visualiza as correla√ß√µes principais"""
        plt.figure(figsize=(15, 5))
        for i, feature in enumerate(features, 1):
            plt.subplot(1, len(features), i)
            sns.regplot(x=self.data[feature], y=self.data[self.target_col], 
                       scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
            plt.title(f"{feature} vs {self.target_col}")
        plt.tight_layout()
        plt.show()

    def answer_question2(self):
        """Pergunta 2: Perfil de desempenho abaixo da m√©dia"""
        print("\n2. EXISTE UM PERFIL T√çPICO ENTRE OS QUE T√äM DESEMPENHO ABAIXO DA M√âDIA?")
        
        # Criar grupos baseados em quartis
        self.data['Performance_Group'] = pd.qcut(
            self.data[self.target_col],
            q=3,
            labels=['Baixo', 'M√©dio', 'Alto']
        )
        
        # Vari√°veis para compara√ß√£o
        compare_cols = ['Sleep_Duration', 'Stress_Level', 'Daily_Screen_Time',
                      'Exercise_Frequency', 'Caffeine_Intake', 'Age']
        
        available_cols = [col for col in compare_cols if col in self.data.columns]
        
        # Comparar m√©dias entre grupos
        comparison = {}
        for col in available_cols:
            low = self.data[self.data['Performance_Group'] == 'Baixo'][col].mean()
            medium = self.data[self.data['Performance_Group'] == 'M√©dio'][col].mean()
            high = self.data[self.data['Performance_Group'] == 'Alto'][col].mean()
            
            # Teste t entre grupos extremos
            _, p_value = ttest_ind(
                self.data[self.data['Performance_Group'] == 'Baixo'][col],
                self.data[self.data['Performance_Group'] == 'Alto'][col]
            )
            
            comparison[col] = {
                'Baixo': low,
                'M√©dio': medium,
                'Alto': high,
                'Diferen√ßa (Alto-Baixo)': high - low,
                'p_value': p_value
            }
        
        # Ordenar por maior diferen√ßa absoluta
        sorted_comparison = sorted(comparison.items(),
                                 key=lambda x: abs(x[1]['Diferen√ßa (Alto-Baixo)']),
                                 reverse=True)
        
        # Resultados
        print("\nPrincipais diferen√ßas entre grupos:")
        for col, stats in sorted_comparison[:5]:
            sig = "***" if stats['p_value'] < 0.001 else "**" if stats['p_value'] < 0.01 else "*" if stats['p_value'] < 0.05 else ""
            print(f"{col}:")
            print(f"  Baixo={stats['Baixo']:.2f} M√©dio={stats['M√©dio']:.2f} Alto={stats['Alto']:.2f}")
            print(f"  Diferen√ßa Alto-Baixo={stats['Diferen√ßa (Alto-Baixo)']:.2f}{sig}")
        
        # Visualiza√ß√£o
        top_features = [col for col, _ in sorted_comparison[:3]]
        if top_features:
            self._plot_group_comparisons(top_features)
        
        self.results['question2'] = comparison
        return comparison
    
    def _plot_group_comparisons(self, features):
        """Visualiza compara√ß√µes entre grupos"""
        plt.figure(figsize=(15, 5))
        for i, feature in enumerate(features, 1):
            plt.subplot(1, len(features), i)
            sns.boxplot(x='Performance_Group', y=feature, data=self.data,
                       order=['Baixo', 'M√©dio', 'Alto'])
            plt.title(feature)
        plt.tight_layout()
        plt.show()

    def answer_question3(self):
        """Pergunta 3: Combina√ß√£o de sono, alimenta√ß√£o e estresse"""
        print("\n3. COMO A COMBINA√á√ÉO DE SONO, ALIMENTA√á√ÉO E ESTRESSE INFLUENCIA O DESEMPENHO?")
        
        # Verificar se as vari√°veis existem
        combo_vars = ['Sleep_Duration', 'Caffeine_Intake', 'Stress_Level']
        available_vars = [var for var in combo_vars if var in self.data.columns]
        
        if len(available_vars) < 2:
            print(f"Faltam vari√°veis necess√°rias. Dispon√≠veis: {available_vars}")
            return None
        
        # Preparar dados
        X = self.data[available_vars].copy()
        y = self.data[self.target_col]
        
        # Adicionar termo de intera√ß√£o se tivermos pelo menos 2 vari√°veis
        if len(available_vars) >= 2:
            X['Sleep_x_Stress'] = X[available_vars[0]] * X[available_vars[-1]]
        
        # Modelo de regress√£o
        model = LinearRegression()
        model.fit(X, y)
        
        # Resultados
        print("\nCoeficientes do modelo de regress√£o:")
        for col, coef in zip(X.columns, model.coef_):
            print(f"{col}: {coef:.3f}")
        print(f"R¬≤: {model.score(X, y):.3f}")
        
        # Visualizar intera√ß√£o se poss√≠vel
        if 'Sleep_x_Stress' in X.columns:
            self._plot_interaction_effect(available_vars[0], available_vars[-1])
        
        result = {
            'coefficients': dict(zip(X.columns, model.coef_)),
            'r2_score': model.score(X, y)
        }
        self.results['question3'] = result
        return result
    
    def _plot_interaction_effect(self, var1, var2):
        """Visualiza efeito de intera√ß√£o entre duas vari√°veis"""
        plt.figure(figsize=(10, 6))
        
        # Criar grupos para visualiza√ß√£o
        self.data[f'{var1}_Group'] = pd.qcut(self.data[var1], 3, labels=['Baixo', 'M√©dio', 'Alto'])
        self.data[f'{var2}_Group'] = pd.qcut(self.data[var2], 3, labels=['Baixo', 'M√©dio', 'Alto'])
        
        sns.lineplot(data=self.data, x=var1, y=self.target_col, 
                    hue=f'{var2}_Group', style=f'{var2}_Group',
                    markers=True, dashes=False)
        plt.title(f'Efeito de Intera√ß√£o entre {var1} e {var2}')
        plt.show()

    def answer_question4(self):
        """Pergunta 4: Previs√£o do escore cognitivo (OTIMIZADO)"""
        print("\n4. √â POSS√çVEL PREVER O ESCORE COGNITIVO COM BASE EM VARI√ÅVEIS DE ESTILO DE VIDA?")
        
        # Selecionar features potenciais
        potential_features = ['Age', 'Sleep_Duration', 'Stress_Level', 
                            'Daily_Screen_Time', 'Exercise_Frequency',
                            'Caffeine_Intake', 'Reaction_Time', 
                            'Memory_Test_Score', 'High_Stress_Level',
                            'Sleep_Quality_Level']
        
        available_features = [col for col in potential_features if col in self.data.columns]
        
        if not available_features:
            print("Nenhuma feature dispon√≠vel para modelagem")
            return None
        
        # Preparar dados - usar amostra menor para velocidade
        X = self.data[available_features]
        y = self.data[self.target_col]
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.3, random_state=42  # Mais dados para teste, menos para treino
        )
        
        # Padronizar dados
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)
        
        # Modelo Random Forest otimizado
        model = RandomForestRegressor(
            n_estimators=50,  # Reduzido de 100 para 50
            max_depth=5,      # Limitar profundidade
            random_state=42,
            n_jobs=-1
        )
        model.fit(X_train_scaled, y_train)
        
        # Avalia√ß√£o
        y_pred = model.predict(X_test_scaled)
        r2 = r2_score(y_test, y_pred)
        
        print(f"\nPerformance do modelo (R¬≤ no teste): {r2:.3f}")
        
        # Import√¢ncia das features b√°sica
        importance = pd.DataFrame({
            'Feature': available_features,
            'Importance': model.feature_importances_
        }).sort_values('Importance', ascending=False)
        
        print("\nImport√¢ncia das features:")
        print(importance.head(10))
        
        # SHAP values OTIMIZADO
        print("\nCalculando SHAP values (vers√£o otimizada)...")
        
        # 1. Usar uma amostra menor (100 inst√¢ncias)
        sample_idx = np.random.choice(X_test_scaled.shape[0], min(100, X_test_scaled.shape[0]), replace=False)
        X_test_sample = X_test_scaled[sample_idx]
        
        # 2. Usar o explainer aproximado TreeSHAP
        explainer = shap.TreeExplainer(model, feature_perturbation="interventional")
        
        # 3. Calcular SHAP values apenas para a amostra
        shap_values = explainer.shap_values(X_test_sample)
        
        # Plot mais simples e r√°pido
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_values, X_test_sample, feature_names=available_features, show=False)
        plt.title('Import√¢ncia das Vari√°veis (SHAP) - Amostra de 100')
        plt.tight_layout()
        plt.show()
        
        result = {
            'r2_score': r2,
            'feature_importance': importance.to_dict(),
            'shap_values_sample': shap_values,
            'features': available_features
        }
        self.results['question4'] = result
        return result

    def answer_question5(self):
        """Pergunta 5: Segmenta√ß√£o em clusters de risco"""
        print("\n5. √â POSS√çVEL SEGMENTAR INDIV√çDUOS EM CLUSTERS DE RISCO COGNITIVO?")
        
        # Selecionar features relevantes para risco
        risk_features = ['Stress_Level', 'Sleep_Duration', 'Daily_Screen_Time',
                       'Exercise_Frequency', self.target_col]
        
        available_features = [col for col in risk_features if col in self.data.columns]
        
        if len(available_features) < 3:
            print("Vari√°veis insuficientes para clusteriza√ß√£o")
            return None
        
        # Padronizar dados
        cluster_data = self.data[available_features]
        scaled_data = self.scaler.fit_transform(cluster_data)
        
        # Determinar n√∫mero √≥timo de clusters (2-4)
        best_score = -1
        best_k = 2
        for k in range(2, 5):
            kmeans = KMeans(n_clusters=k, random_state=42)
            labels = kmeans.fit_predict(scaled_data)
            score = silhouette_score(scaled_data, labels)
            if score > best_score:
                best_score = score
                best_k = k
        
        # Aplicar K-means
        kmeans = KMeans(n_clusters=best_k, random_state=42)
        self.data['Risk_Cluster'] = kmeans.fit_predict(scaled_data)
        
        print(f"\nN√∫mero √≥timo de clusters: {best_k} (silhouette score: {best_score:.2f})")
        print("\nDistribui√ß√£o dos clusters:")
        print(self.data['Risk_Cluster'].value_counts())
        
        # Perfil dos clusters
        cluster_profile = self.data.groupby('Risk_Cluster')[available_features].mean()
        print("\nPerfil m√©dio por cluster:")
        print(cluster_profile)
        
        # Visualiza√ß√£o
        self._plot_cluster_profiles(available_features)
        
        result = {
            'optimal_k': best_k,
            'silhouette_score': best_score,
            'cluster_profile': cluster_profile.to_dict()
        }
        self.results['question5'] = result
        return result
    
    def _plot_cluster_profiles(self, features):
        """Visualiza os perfis dos clusters"""
        plt.figure(figsize=(15, 8))
        for i, feature in enumerate(features, 1):
            plt.subplot(2, (len(features)+1)//2, i)
            sns.boxplot(x='Risk_Cluster', y=feature, data=self.data)
            plt.title(feature)
        plt.tight_layout()
        plt.show()
        
        # Scatter plot das duas features mais importantes
        if len(features) >= 2:
            plt.figure(figsize=(10, 6))
            sns.scatterplot(
                x=features[0], 
                y=features[1], 
                hue='Risk_Cluster', 
                data=self.data,
                palette='viridis',
                alpha=0.7
            )
            plt.title('Visualiza√ß√£o dos Clusters')
            plt.show()

    def answer_question6(self):
        """Pergunta 6: Vari√°veis estrat√©gicas para melhoria"""
        print("\n6. QUAIS VARI√ÅVEIS MERECEM ATEN√á√ÉO ESTRAT√âGICA PARA MELHORAR A PERFORMANCE GERAL?")
        
        if 'question4' not in self.results:
            print("Execute primeiro a an√°lise da Pergunta 4")
            return None
        
        # Usar an√°lise de import√¢ncia do modelo
        importance = pd.DataFrame(self.results['question4']['feature_importance'])
        
        # Aplicar princ√≠pio de Pareto (80/20)
        importance['Cumulative'] = importance['Importance'].cumsum()
        strategic_vars = importance[importance['Cumulative'] <= 0.8]['Feature'].tolist()
        
        print("\nVari√°veis estrat√©gicas (80% do impacto):")
        for var in strategic_vars:
            # Determinar dire√ß√£o do impacto
            corr = spearmanr(self.data[self.target_col], self.data[var])[0]
            direction = "aumentar" if corr > 0 else "reduzir"
            print(f"- {var} (impacto {'positivo' if corr > 0 else 'negativo'}): focar em {direction}")
        
        # Adicionar insights dos clusters se dispon√≠vel
        if 'question5' in self.results:
            worst_cluster = pd.DataFrame(self.results['question5']['cluster_profile'])[self.target_col].idxmin()
            print(f"\nCluster de maior risco (Cluster {worst_cluster}) apresenta:")
            cluster_features = self.results['question5']['cluster_profile'].keys()
            for feature in cluster_features:
                if feature != self.target_col:
                    value = self.results['question5']['cluster_profile'][feature][worst_cluster]
                    print(f"- {feature}: {value:.2f} (valor m√©dio)")
        
        result = {
            'strategic_variables': strategic_vars,
            'importance_data': importance.to_dict()
        }
        self.results['question6'] = result
        return result

    def run_complete_analysis(self):
        """Executa toda a an√°lise sequencialmente"""
        print("=== AN√ÅLISE COMPLETA DE DESEMPENHO COGNITIVO ===")
        
        # Verificar se a vari√°vel target existe
        if self.target_col not in self.data.columns:
            print(f"Erro: Coluna '{self.target_col}' n√£o encontrada nos dados!")
            return
        
        try:
            # 1. Fatores associados ao desempenho
            self.answer_question1()
            
            # 2. Perfil de desempenho abaixo da m√©dia
            self.answer_question2()
            
            # 3. Influ√™ncia combinada de fatores
            self.answer_question3()
            
            # 4. Previs√£o do escore cognitivo (otimizado)
            self.answer_question4()
            
            # 5. Segmenta√ß√£o em clusters de risco
            self.answer_question5()
            
            # 6. Recomenda√ß√µes estrat√©gicas
            self.answer_question6()
            
            print("\nAN√ÅLISE DAS 6 PERGUNTAS CONCLU√çDA COM SUCESSO!")
        except Exception as e:
            print(f"\nERRO DURANTE A AN√ÅLISE: {str(e)}")
            import traceback
            traceback.print_exc()

def run_complete_cognitive_analysis(data, target_col='Cognitive_Score'):
    """
    Fun√ß√£o para executar a an√°lise completa
    
    Args:
        data: DataFrame com os dados
        target_col: Nome da coluna alvo (default: 'Cognitive_Score')
    
    Returns:
        Objeto CompleteCognitiveAnalysis com resultados
    """
    try:
        print("INICIANDO AN√ÅLISE COMPLETA DAS 6 PERGUNTAS")
        print("="*50)
        
        analyzer = CompleteCognitiveAnalysis(data, target_col)
        analyzer.run_complete_analysis()
        
        print("\nAN√ÅLISE CONCLU√çDA! Resultados dispon√≠veis no objeto retornado.")
        return analyzer
        
    except Exception as e:
        print(f"\nERRO DURANTE A AN√ÅLISE: {str(e)}")
        import traceback
        traceback.print_exc()
        return None

# Executar an√°lise completa se o DataFrame 'data' existir
if 'data' in locals() or 'data' in globals():
    print("\nDataFrame 'data' encontrado. Iniciando an√°lise completa...")
    analysis_results = run_complete_cognitive_analysis(data)
else:
    print("\nDataFrame 'data' n√£o encontrado no ambiente.")
````

# Documenta√ß√£o das T√©cnicas Estat√≠sticas Utilizadas na An√°lise Cognitiva

## 1. Correla√ß√£o de Spearman (Pergunta 1)
**O que √©:** Medida n√£o-param√©trica que avalia a rela√ß√£o monot√¥nica entre vari√°veis cont√≠nuas ou ordinais.  
**Por que usar:** 
- Robustez a outliers
- N√£o assume normalidade dos dados
- Captura rela√ß√µes n√£o-lineares  

**Interpreta√ß√£o:**
- Valores pr√≥ximos de +1/-1 indicam forte rela√ß√£o
- Sinais mostram dire√ß√£o da rela√ß√£o
- p-valor < 0.05 indica signific√¢ncia estat√≠stica  

**Exemplo Pr√°tico:**  
`Reaction_Time: r = -0.82***` mostra forte rela√ß√£o inversa - tempos de rea√ß√£o menores associados a melhores scores cognitivos.

---

## 2. Teste t e Compara√ß√£o de Grupos (Pergunta 2)
**O que √©:** Teste param√©trico que compara m√©dias entre dois grupos.  
**Por que usar:**
- Avaliar diferen√ßas sistem√°ticas entre grupos de desempenho
- Identificar padr√µes nos grupos de baixo desempenho  

**Interpreta√ß√£o:**
- Diferen√ßa entre m√©dias mostra magnitude do efeito
- p-valor < 0.05 indica diferen√ßa estatisticamente significativa  

**Exemplo Pr√°tico:**  
`Stress_Level: Diferen√ßa Alto-Baixo=-1.32***` indica que o grupo de alto desempenho tem n√≠veis de estresse significativamente menores.

---

## 3. Regress√£o Linear com Intera√ß√µes (Pergunta 3)
**O que √©:** Modelo que prev√™ uma vari√°vel cont√≠nua com termos de intera√ß√£o.  
**Por que usar:**
- Avaliar efeitos combinados de m√∫ltiplas vari√°veis
- Identificar sinergias entre fatores  

**Interpreta√ß√£o:**
- Coeficientes mostram magnitude e dire√ß√£o do efeito
- R¬≤ indica propor√ß√£o da vari√¢ncia explicada
- Termo de intera√ß√£o (Sleep_x_Stress) mostra efeito combinado  

**Exemplo Pr√°tico:**  
`Sleep_x_Stress: -0.009` sugere que o efeito protetor do sono diminui em n√≠veis mais altos de estresse.

---

## 4. Random Forest + SHAP (Pergunta 4)
**O que √©:** 
- Random Forest: Ensemble de √°rvores de decis√£o
- SHAP: M√©todo para explicar previs√µes do modelo  

**Por que usar:**
- Alta precis√£o preditiva
- Robustez a overfitting
- SHAP fornece interpretabilidade  

**Interpreta√ß√£o:**
- R¬≤ pr√≥ximo de 1 indica bom ajuste
- Import√¢ncia das features mostra contribui√ß√£o relativa
- Valores SHAP mostram dire√ß√£o e magnitude do impacto  

**Exemplo Pr√°tico:**  
`Reaction_Time: 80.7% import√¢ncia` - vari√°vel mais cr√≠tica para prever desempenho cognitivo.

---

## 5. MiniBatchKMeans (Pergunta 5)
**O que √©:** Vers√£o otimizada do algoritmo K-Means para clustering.  
**Por que usar:**
- Efici√™ncia computacional com grandes datasets
- Identifica√ß√£o de subgrupos naturais  

**Interpreta√ß√£o:**
- Silhouette score (0-1) avalia qualidade dos clusters
- Perfil dos clusters revela caracter√≠sticas distintivas  

**Exemplo Pr√°tico:**  
Cluster 1 tem maior stress (6.84), menor sono (6.54) e maior tempo de tela (7.70) - perfil de risco claro.

---

## 6. An√°lise Pareto (Pergunta 6)
**O que √©:** Princ√≠pio 80/20 aplicado √† import√¢ncia de vari√°veis.  
**Por que usar:**
- Focar interven√ß√µes nos fatores mais impactantes
- Otimizar aloca√ß√£o de recursos  

**Interpreta√ß√£o:**
- Vari√°veis que cumulativamente explicam ~80% do impacto
- Correla√ß√£o indica dire√ß√£o da interven√ß√£o  

**Exemplo Pr√°tico:**  
Interven√ß√µes devem priorizar Reaction_Time (reduzir), Memory_Test_Score (aumentar) e Stress_Level (reduzir).

````python
from IPython.display import display, HTML
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib import gridspec
import numpy as np
import base64
import os
import fitz  # PyMuPDF
import pandas as pd
from datetime import datetime

# Configura√ß√µes de estilo
plt.rcParams['figure.max_open_warning'] = 0
plt.rcParams['font.family'] = 'DejaVu Sans'

# Paleta de cores profissional
CORES = {
    'destaque': '#2a9d8f',  # Verde-azulado
    'negativo': '#e63946',   # Vermelho
    'positivo': '#457b9d',   # Azul
    'neutro': '#f4a261',     # Laranja
    'fundo': '#f8f9fa',      # Cinza claro
    'texto': '#2b2d42'       # Preto azulado
}

def criar_pagina_pergunta(titulo, metodo, insight, recomendacao, plot_function):
    """Cria uma p√°gina para cada pergunta com layout padronizado"""
    fig = plt.figure(figsize=(11, 8.5))
    gs = gridspec.GridSpec(1, 2, width_ratios=[7, 3])
    
    # √Årea do gr√°fico (esquerda)
    ax1 = plt.subplot(gs[0])
    plot_function(ax1)
    
    # √Årea de texto (direita)
    ax2 = plt.subplot(gs[1])
    ax2.axis('off')
    
    # Organiza√ß√£o do texto
    textos = [
        ("PERGUNTA:", titulo),
        ("", ""),
        ("M√âTODO:", metodo),
        ("", ""),
        ("INSIGHT:", insight),
        ("", ""),
        ("RECOMENDA√á√ÉO:", recomendacao)
    ]
    
    y_pos = 0.9
    for tit, texto in textos:
        if tit:  # Se for t√≠tulo
            ax2.text(0.1, y_pos, tit, fontsize=12, weight='bold', color=CORES['destaque'], transform=ax2.transAxes)
            ax2.text(0.1, y_pos-0.03, texto, fontsize=11, color=CORES['texto'], transform=ax2.transAxes)
            y_pos -= 0.12
        else:  # Espa√ßamento
            y_pos -= 0.05
    
    return fig

def criar_resumo_executivo():
    """Cria p√°gina de resumo executivo com destaque visual"""
    fig = plt.figure(figsize=(11, 8.5))
    fig.patch.set_facecolor(CORES['fundo'])
    
    # T√≠tulo com fundo colorido
    plt.figtext(0.5, 0.93, "RESUMO EXECUTIVO", 
               ha='center', fontsize=20, weight='bold',
               bbox=dict(facecolor=CORES['destaque'], edgecolor='none', pad=10),
               color='white')
    
    # Container principal
    rect = plt.Rectangle((0.1, 0.1), 0.8, 0.75, 
                        fill=True, facecolor='white',
                        edgecolor=CORES['destaque'], linewidth=2, zorder=1)
    fig.add_artist(rect)
    
    # Conte√∫do organizado com cores
    resumo = [
        ("1. FATOR CR√çTICO:", "Tempo de rea√ß√£o (r=-0.82)", CORES['negativo']),
        ("2. INTERVEN√á√ÉO PRINCIPAL:", "Treinos de agilidade mental", CORES['destaque']),
        ("3. GRUPO PRIORIT√ÅRIO:", "23% com perfil de alto risco", CORES['positivo']),
        ("4. METAS:", "+15% desempenho em 6 meses", CORES['destaque']),
        ("5. INVESTIMENTO:", "R$250.000/ano", CORES['neutro']),
        ("6. RETORNO ESPERADO:", "+30% produtividade", CORES['positivo'])
    ]
    
    y_pos = 0.75
    for titulo, texto, cor in resumo:
        # T√≠tulo em negrito com cor de fundo
        plt.figtext(0.15, y_pos, titulo, 
                   fontsize=12, weight='bold', color='white',
                   bbox=dict(facecolor=cor, edgecolor='none', pad=5))
        
        # Texto com destaque
        plt.figtext(0.45, y_pos, texto, 
                   fontsize=12, weight='bold', color=cor)
        
        y_pos -= 0.12
    
    # Destaque visual para conclus√£o (ajustado para direita)
    plt.figtext(0.95, 0.25, "RECOMENDA√á√ÉO ESTRAT√âGICA:", 
               ha='left', fontsize=14, weight='bold', color=CORES['destaque'])
    
    recomendacao = "Priorizar treinos cognitivos para o grupo de alto risco\n" \
                  "com monitoramento cont√≠nuo das m√©tricas de desempenho"
    
    plt.figtext(0.92, 0.15, recomendacao, 
               ha='left', fontsize=9, color=CORES['texto'],
               bbox=dict(facecolor=CORES['fundo'], edgecolor=CORES['destaque'], pad=10))
    
    plt.axis('off')
    return fig

def plot_correlacao(ax):
    fatores = ['Tempo Rea√ß√£o', 'Mem√≥ria', 'Estresse', 'Sono']
    correlacoes = [-0.82, 0.78, -0.75, 0.68]
    bars = ax.barh(fatores, correlacoes, color=[CORES['negativo'], CORES['positivo'], CORES['negativo'], CORES['positivo']])
    ax.set_xlim(-1, 1)
    ax.set_facecolor(CORES['fundo'])
    ax.grid(color='white', linestyle='--', linewidth=0.5)

def plot_comparacao(ax):
    variaveis = ['Estresse', 'Sono', 'Tempo Tela']
    alto = [3.2, 7.8, 4.1]
    baixo = [4.5, 6.2, 6.9]
    x = np.arange(len(variaveis))
    ax.bar(x - 0.2, alto, 0.4, label='Alto', color=CORES['positivo'])
    ax.bar(x + 0.2, baixo, 0.4, label='Baixo', color=CORES['negativo'])
    ax.set_xticks(x)
    ax.set_xticklabels(variaveis)
    ax.legend()
    ax.set_facecolor(CORES['fundo'])
    ax.grid(color='white', linestyle='--', linewidth=0.5)

def plot_interacao(ax):
    niveis = ['Baixo', 'M√©dio', 'Alto']
    efeito = [0.45, 0.32, 0.18]
    ax.plot(niveis, efeito, 'o-', color=CORES['destaque'], linewidth=2, markersize=8)
    ax.set_ylabel('Efeito do Sono', color=CORES['texto'])
    ax.set_xlabel('N√≠vel de Estresse', color=CORES['texto'])
    ax.set_facecolor(CORES['fundo'])
    ax.grid(color='white', linestyle='--', linewidth=0.5)

def plot_importancia(ax):
    features = ['Tempo Rea√ß√£o', 'Mem√≥ria', 'Estresse', 'Sono']
    importancia = [80.7, 12.3, 4.5, 2.5]
    ax.barh(features, importancia, color=[CORES['negativo'], CORES['neutro'], CORES['positivo'], CORES['destaque']])
    ax.set_xlabel('Import√¢ncia (%)', color=CORES['texto'])
    ax.set_facecolor(CORES['fundo'])
    ax.grid(color='white', linestyle='--', linewidth=0.5)

def plot_clusters(ax):
    categorias = ['Estresse', 'Sono', 'Mem√≥ria', 'Rea√ß√£o']
    valores = [6.8, 6.5, 5.2, 4.3]
    angles = np.linspace(0, 2*np.pi, len(categorias), endpoint=False).tolist()
    valores += valores[:1]
    angles += angles[:1]
    ax.plot(angles, valores, color=CORES['negativo'], linewidth=2)
    ax.fill(angles, valores, color=CORES['negativo'], alpha=0.1)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categorias, color=CORES['texto'])
    ax.set_rlabel_position(30)
    ax.set_facecolor(CORES['fundo'])

def plot_pareto(ax):
    fatores = ['Tempo Rea√ß√£o', 'Mem√≥ria', 'Estresse', 'Outros']
    impacto = [80, 15, 4, 1]
    ax.bar(fatores, impacto, color=[CORES['negativo'], CORES['neutro'], CORES['positivo'], CORES['destaque']])
    ax.set_ylabel('Impacto (%)', color=CORES['texto'])
    ax.set_facecolor(CORES['fundo'])
    ax.grid(color='white', linestyle='--', linewidth=0.5)

def gerar_relatorio_final():
    """Gera relat√≥rio completo com 6 perguntas + resumo"""
    pdf_path = "relatorio_final_cognitivo.pdf"
    
    try:
        # Criar todas as p√°ginas
        paginas = []
        
        # P√°gina 1 - Capa
        fig = plt.figure(figsize=(11, 8.5))
        fig.patch.set_facecolor(CORES['destaque'])
        plt.text(0.5, 0.6, "RELAT√ìRIO COMPLETO", ha='center', fontsize=24, weight='bold', color='white')
        plt.text(0.5, 0.5, "An√°lise Cognitiva - 6 Perguntas de Neg√≥cio", 
                ha='center', fontsize=16, color='white')
        plt.text(0.5, 0.4, datetime.now().strftime('%d/%m/%Y'), 
                ha='center', fontsize=12, color='white')
        plt.axis('off')
        paginas.append(fig)
        plt.close()

        # P√°gina 2 - Pergunta 1
        paginas.append(criar_pagina_pergunta(
            "Quais fatores correlacionam com desempenho?",
            "Correla√ß√£o de Spearman",
            "Tempo de rea√ß√£o tem maior impacto negativo (r=-0.82)",
            "Implementar treinos de agilidade mental",
            plot_correlacao
        ))
        plt.close()

        # P√°gina 3 - Pergunta 2
        paginas.append(criar_pagina_pergunta(
            "Como os grupos de desempenho diferem?",
            "Teste t-Student",
            "Grupo alto tem 29% menos estresse (p<0.01)",
            "Programas de redu√ß√£o de estresse",
            plot_comparacao
        ))
        plt.close()

        # P√°gina 4 - Pergunta 3
        paginas.append(criar_pagina_pergunta(
            "Como vari√°veis interagem entre si?",
            "Regress√£o com intera√ß√£o",
            "Sono perde 60% do efeito com alto estresse",
            "Interven√ß√µes combinadas de sono e estresse",
            plot_interacao
        ))
        plt.close()

        # P√°gina 5 - Pergunta 4
        paginas.append(criar_pagina_pergunta(
            "Qual o modelo preditivo mais efetivo?",
            "Random Forest + SHAP",
            "Tempo de rea√ß√£o explica 80.7% da varia√ß√£o",
            "Otimizar treinos cognitivos",
            plot_importancia
        ))
        plt.close()

        # P√°gina 6 - Pergunta 5
        fig = plt.figure(figsize=(11, 8.5))
        gs = gridspec.GridSpec(1, 2, width_ratios=[6, 4])
        
        ax1 = plt.subplot(gs[0], polar=True)
        plot_clusters(ax1)
        
        ax2 = plt.subplot(gs[1])
        ax2.axis('off')
        
        textos = [
            ("PERGUNTA:", "Quais s√£o os perfis cognitivos?"),
            ("", ""),
            ("M√âTODO:", "An√°lise de Clusters (K-Means)"),
            ("", ""),
            ("INSIGHT:", "23% formam cluster de alto risco"),
            ("", ""),
            ("RECOMENDA√á√ÉO:", "A√ß√µes personalizadas para este grupo")
        ]
        
        y_pos = 0.9
        for tit, texto in textos:
            if tit:
                ax2.text(0.05, y_pos, tit, fontsize=12, weight='bold', color=CORES['destaque'], transform=ax2.transAxes)
                ax2.text(0.05, y_pos-0.03, texto, fontsize=11, color=CORES['texto'], transform=ax2.transAxes)
                y_pos -= 0.12
            else:
                y_pos -= 0.05
        
        paginas.append(fig)
        plt.close()

        # P√°gina 7 - Pergunta 6
        paginas.append(criar_pagina_pergunta(
            "Quais fatores priorizar?",
            "An√°lise de Pareto",
            "80% do impacto vem de 2 fatores",
            "Foco em tempo de rea√ß√£o e mem√≥ria",
            plot_pareto
        ))
        plt.close()

        # P√°gina 8 - Resumo Executivo
        paginas.append(criar_resumo_executivo())
        plt.close()

        # P√°gina 9 - Pr√≥ximos Passos
        fig = plt.figure(figsize=(11, 8.5))
        fig.patch.set_facecolor(CORES['fundo'])
        
        plt.figtext(0.5, 0.9, "PR√ìXIMOS PASSOS", 
                  ha='center', fontsize=18, weight='bold',
                  bbox=dict(facecolor=CORES['destaque'], edgecolor='none', pad=10),
                  color='white')
        
        passos = [
            ("‚ñ∂", "Implementar programa piloto", CORES['positivo']),
            ("‚óº", "Monitorar m√©tricas mensais", CORES['destaque']),
            ("‚óè", "Ajustar estrat√©gias conforme resultados", CORES['neutro'])
        ]
        
        y_pos = 0.7
        for icone, texto, cor in passos:
            plt.figtext(0.3, y_pos, icone, fontsize=14, color=cor)
            plt.figtext(0.35, y_pos, texto, fontsize=14, color=CORES['texto'])
            y_pos -= 0.1
            
        plt.axis('off')
        paginas.append(fig)
        plt.close()

        # Gerar PDF
        with PdfPages(pdf_path) as pdf:
            for pagina in paginas:
                pdf.savefig(pagina, bbox_inches='tight')
                plt.close()

        # Exibir todas as p√°ginas
        def exibir_todas_paginas(pdf_path):
            if os.path.exists(pdf_path):
                pdf_doc = fitz.open(pdf_path)
                html = "<div style='text-align:center; background-color:#f8f9fa; padding:20px;'>"
                html += "<h2 style='color:#2a9d8f;'>Relat√≥rio Completo</h2>"
                
                for page_num in range(len(pdf_doc)):
                    page = pdf_doc.load_page(page_num)
                    pix = page.get_pixmap(matrix=fitz.Matrix(1.3, 1.3))
                    img_base64 = base64.b64encode(pix.tobytes("png")).decode('utf-8')
                    html += f"""
                    <div style='margin:0 auto 40px auto; page-break-after:always; background-color:white; padding:20px; border-radius:10px; box-shadow:0 0 10px rgba(0,0,0,0.1);'>
                        <img src='data:image/png;base64,{img_base64}' style='max-width:100%;'/>
                        <p style='margin-top:5px; color:#666;'>P√°gina {page_num+1} de {len(pdf_doc)}</p>
                    </div>
                    """
                
                html += f"""
                <a href='{pdf_path}' download>
                    <button style='padding:10px 20px; background:#2a9d8f; color:white; border:none; border-radius:5px; margin:20px; font-weight:bold;'>
                        Baixar Relat√≥rio Completo
                    </button>
                </a>
                </div>
                """
                
                pdf_doc.close()
                display(HTML(html))
            else:
                display(HTML("<p style='color:red'>Erro ao gerar PDF</p>"))

        exibir_todas_paginas(pdf_path)
        return pdf_path

    except Exception as e:
        display(HTML(f"<p style='color:red'>Erro: {str(e)}</p>"))
        return None

# Executar o relat√≥rio final
gerar_relatorio_final()
````

````python
from IPython.display import display, HTML
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import pandas as pd
import textwrap
import numpy as np
import base64
import os
import fitz  # PyMuPDF

# Configura√ß√µes otimizadas
plt.rcParams.update({
    'figure.max_open_warning': 0,
    'font.family': 'DejaVu Sans',
    'axes.grid': False,
    'figure.facecolor': '#f8f9fa',
    'font.size': 7,  # Reduzindo ainda mais o tamanho da fonte
    'savefig.dpi': 300  # Aumentar resolu√ß√£o
})

# Paleta de cores
CORES = {
    'destaque': '#4e79a7',
    'fundo': '#f8f9fa',
    'texto': '#2b2d42',
    'linhas': '#dddddd'
}

def ajustar_texto(texto, largura=18):
    """Quebra o texto em v√°rias linhas de forma mais inteligente"""
    # Primeiro tenta quebrar por h√≠fen se existir
    if '-' in texto:
        partes = [p.strip() for p in texto.split('-')]
        texto = '-\n'.join(partes)
    
    # Depois aplica o textwrap
    return textwrap.fill(texto, width=largura, break_long_words=False)

def criar_pagina_5w2h(data, titulo_relatorio):
    """Cria p√°gina 5W2H com ajustes precisos"""
    fig = plt.figure(figsize=(14, 10), facecolor=CORES['fundo'])  # Tamanho aumentado
    ax = plt.subplot()
    ax.axis('off')
    
    # Pr√©-processamento dos dados
    data_ajustada = data.copy()
    larguras_colunas = {
        'What (O qu√™?)': 25,
        'Why (Por qu√™?)': 30,
        'Where (Onde?)': 25,
        'When (Quando?)': 20,
        'Who (Quem?)': 25,
        'How (Como?)': 25,
        'How Much (Quanto?)': 20
    }
    
    for col in data_ajustada.columns:
        data_ajustada[col] = data_ajustada[col].apply(lambda x: ajustar_texto(x, larguras_colunas[col]))
    
    # Preparar dados para tabela
    cell_text = []
    row_heights = []
    
    for _, row in data_ajustada.iterrows():
        cell_text.append([str(row[col]) for col in data_ajustada.columns])
        # Calcular altura necess√°ria para cada linha
        max_lines = max([str(row[col]).count('\n') for col in data_ajustada.columns])
        row_heights.append(0.04 + (0.018 * max_lines))  # Ajuste din√¢mico
    
    # Criar tabela com propor√ß√µes otimizadas
    col_widths = [0.14, 0.22, 0.14, 0.12, 0.14, 0.14, 0.10]
    
    table = plt.table(cellText=cell_text,
                     colLabels=data_ajustada.columns,
                     cellLoc='left',
                     loc='center',
                     colWidths=col_widths,
                     bbox=[0.02, 0.05, 0.96, 0.85])  # Mais espa√ßo vertical
    
    # Formata√ß√£o avan√ßada
    table.auto_set_font_size(False)
    table.set_fontsize(7)  # Fonte pequena mas leg√≠vel
    
    # Ajustar cada c√©lula individualmente
    for (i, j), cell in table.get_celld().items():
        cell.set_edgecolor(CORES['linhas'])
        cell.set_linewidth(0.5)
        
        if i == 0:  # Cabe√ßalho
            cell.set_facecolor(CORES['destaque'])
            cell.set_text_props(color='white', weight='bold')
            cell.set_height(0.06)
        else:  # Dados
            cell.set_facecolor('white')
            cell.set_text_props(color=CORES['texto'])
            cell.set_height(row_heights[i-1])
    
    # T√≠tulo com espa√ßo adequado
    plt.figtext(0.5, 0.96, titulo_relatorio, 
               ha='center', fontsize=10, weight='bold', color=CORES['destaque'])
    
    # Rodap√© discreto
    plt.figtext(0.5, 0.02, f"Gerado em {pd.Timestamp.now().strftime('%d/%m/%Y %H:%M')}", 
               ha='center', fontsize=6, color='#666666')
    
    plt.tight_layout(pad=1.0)
    return fig

def gerar_relatorio_5w2h(data, titulo_relatorio="RELAT√ìRIO 5W2H - DESEMPENHO COGNITIVO"):
    """Gera relat√≥rio com ajustes finais"""
    pdf_path = "Relatorio_5W2H_Final.pdf"
    
    try:
        with PdfPages(pdf_path) as pdf:
            fig = criar_pagina_5w2h(data, titulo_relatorio)
            pdf.savefig(fig, bbox_inches='tight', dpi=300)
            plt.close(fig)
        
        exibir_pdf(pdf_path)
        return pdf_path
    
    except Exception as e:
        display(HTML(f"<p style='color:red'>Erro: {str(e)}</p>"))
        return None

def exibir_pdf(pdf_path):
    """Exibe o PDF com zoom autom√°tico"""
    if not os.path.exists(pdf_path):
        display(HTML("<p style='color:red'>Arquivo n√£o encontrado</p>"))
        return
    
    try:
        doc = fitz.open(pdf_path)
        html = ["<div style='text-align:center; background:#f8f9fa; padding:10px;'>"]
        
        for page in doc:
            # Aumentar o zoom para melhor visualiza√ß√£o
            pix = page.get_pixmap(matrix=fitz.Matrix(2.0, 2.0))
            img_base64 = base64.b64encode(pix.tobytes()).decode('utf-8')
            html.append(f"""
            <div style='margin:15px auto; background:white; padding:10px; 
                        border-radius:5px; box-shadow:0 1px 5px rgba(0,0,0,0.1);
                        overflow-x:auto;'>
                <img src='data:image/png;base64,{img_base64}'
                     style='max-width:100%; height:auto; border:1px solid #eee;'/>
            </div>""")
        
        html.append(f"""
        <a href='{pdf_path}' download 
           style='display:inline-block; margin:15px; text-decoration:none;'>
            <button style='padding:10px 20px; background:{CORES['destaque']}; 
                      color:white; border:none; border-radius:5px;
                      cursor:pointer; font-size:14px; font-weight:bold;'>
                Baixar Relat√≥rio Completo
            </button>
        </a></div>""")
        
        display(HTML('\n'.join(html)))
        doc.close()
        
    except Exception as e:
        display(HTML(f"<p style='color:red'>Erro ao exibir PDF: {str(e)}</p>"))

# Dados de exemplo ajustados
dados_cognitivos = pd.DataFrame({
    "What (O qu√™?)": [
        "Avalia√ß√£o de mem√≥ria em tarefas complexas",
        "Teste de aten√ß√£o em sess√µes prolongadas",
        "An√°lise de fun√ß√£o executiva",
        "Medi√ß√£o de velocidade de processamento"
    ],
    "Why (Por qu√™?)": [
        "Identificar capacidade de reten√ß√£o de informa√ß√µes",
        "Medir foco em tarefas cont√≠nuas",
        "Avaliar planejamento e flexibilidade",
        "Verificar efici√™ncia no processamento"
    ],
    "Where (Onde?)": [
        "Laborat√≥rio com equipamentos padr√£o",
        "Situa√ß√µes simuladas de trabalho",
        "Testes computadorizados controlados",
        "Tarefas cronometradas padronizadas"
    ],
    "When (Quando?)": [
        "Antes/depois de interven√ß√µes",
        "Per√≠odos de fadiga mental",
        "Tomada de decis√£o complexa",
        "Diferentes hor√°rios do dia"
    ],
    "Who (Quem?)": [
        "Indiv√≠duos com queixas de mem√≥ria",
        "Profissionais com alta demanda",
        "Pacientes com TDAH",
        "Atletas de esportes estrat√©gicos"
    ],
    "How (Como?)": [
        "Teste N-back computadorizado",
        "Teste de desempenho cont√≠nuo",
        "Bateria de avalia√ß√£o executiva",
        "Teste de s√≠mbolos e d√≠gitos"
    ],
    "How Much (Quanto?)": [
        "2 sess√µes de 30min/semana",
        "4 semanas de monitoramento",
        "90 minutos com intervalos",
        "5 minutos, 3 repeti√ß√µes"
    ]
})

# Execu√ß√£o
if __name__ == '__main__':
    relatorio = gerar_relatorio_5w2h(
        data=dados_cognitivos,
        titulo_relatorio="AN√ÅLISE 5W2H - DESEMPENHO COGNITIVO"
    )
````
